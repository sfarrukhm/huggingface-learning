```python
# using pipeline in its default setting
from transformers import pipeline
classifier=pipeline('sentiment-analysis') # we are not poviding any model but we can. If no model is provided, pipeline() picks default model relevant task.
classifier("Why are you annoying me?")
```

#### Some currently available `pipeline()` taks:
- feature-extraction (makes the vector representation of a text)
- ner
- question-answering
- sentiment-analysis
- summarization
- text-generation
- translation
- zero-shot-classification

### zero-shot-classification:
- The zero-shot-classification pipeline simplifies this by letting you define custom labels, beyond those of the pre-trained model.
- It is callaed `zero-shot` because you don't need to fine-tune the model on your data to use it. It can automatically adjusts itself to custom labels and provide the probability scores.


```python
classifier=pipeline('zero-shot-classification')
```

```python
query= "Eating an apple a day, keeps the doctor away"
classifier(query,
           candidate_labels=["education","politics",'healthcare'])
```

### Mask filling with `fill-mask`

```python
from transformers import pipeline
unmasker=pipeline("fill-mask")
```

```python
unmasker("Eating an <mask> a day, keeps the doctor away.",
         top_k=3)
```

### Named entity recognition with `ner`

```python
ner=pipeline("ner",grouped_entities=True)
```

We pass the option grouped_entities=True in the pipeline creation function to tell the pipeline to regroup together the parts of the sentence that correspond to the same entity.

```python
ner("My city is Bahawalpur, I study in NUST which is in Islamabad")
```

Now we will manually supply a model pretrained on NER of POS

```python
ner=pipeline("ner",model="QCRI/bert-base-multilingual-cased-pos-english",grouped_entities=True)
```

```python
ner("My city is Bahawalpur, I study in NUST which is in Islamabad")
```
